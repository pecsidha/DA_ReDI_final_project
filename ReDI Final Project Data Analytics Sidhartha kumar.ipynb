### clean up data


```python
nan = sales_df[sales_df.isna().any(axis=1)]
nan.head()
sales_df['postal code'] = sales_df['postal code'].astype(str)
```

```python
sales_df = sales_df.dropna(how='all')
sales_df.head()
```

### Add month column

```python
sales_df['Month'] = sales_df['BestellTag'].str[3:5]
sales_df['Month'] = sales_df['Month'].astype('int32')
```

```python
sales_df.head(50)
```

```python
monthly_sales_quantity = sales_df['Month'].value_counts()
monthly_sales_quantity.head(50)
```

```python
import matplotlib.pyplot as plt
months = range(1,13)
plt.bar(months,monthly_sales_quantity)
plt.xticks(months)
plt.ylabel('no. of quanity sold')
plt.xlabel('Months')
plt.show()
```

```python
location_sales = sales_df['postal code'].value_counts(ascending=True)
location_sales.head()
```
